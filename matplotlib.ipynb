```python

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,9,16,25]

plt.plot(x,y)
```

```python
x = [1,2,3,4,5]
y = [1,2,9,16,25]

# customising the plot 

plt.plot(x,y,color = 'red',linestyle = '--',marker = 'o' )
plt.grid(True)
```

```python
## creating multi plots

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize = (9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color = 'green')
plt.title("Plot1 ")

plt.subplot(2,2,2)
plt.plot(x,y1,color = 'black')
plt.title("Plot2 ")

plt.subplot(2,2,3)
plt.plot(x,y1,color = 'red')
plt.title("Plot3")

plt.subplot(2,2,4)
plt.plot(x,y1,color = 'blue')
plt.title("Plot4")
```

```python
### bar plot 

categories = ['a','b','c','d','e']
values = [5,7,7,3,1]

plt.bar(categories,values,color='purple')
```

```python
### histograms 

data = [1,2,2,3,4,4,5,5,5,5,6,6,7,7,8]

plt.hist(data,bins=8,color = 'orange',edgecolor='black')
```

```python
### craete a scatter plot 

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.scatter(x,y,color = 'blue',marker = '*')
```

```python
# pie chart 

labels = ['a','b','c','d','e','f']
sizes = [30,20,20,10,10,10]
colors = ['gold','green','blue','black','yellow','purple']

plt.pie(sizes,labels =  labels,colors = colors)
```
