```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('linearRegression.csv')
df.head()
```

```python
%matplotlib inline
plt.scatter(df.area,df.price)
```

```python
rg = linear_model.LinearRegression()
rg.fit(df[['area']],df.price)
```

```python
rg.predict([[3300]])
```

```python
rg.coef_
```

```python
rg.intercept_
```

```python
135.78767123*3300 + 180616.43835616432
```

```python
%matplotlib inline
plt.scatter(df.area,df.price)
plt.plot(df.area,rg.predict(df[['area']]),color='red')
plt.xlabel('Area')
plt.ylabel('Price')
```
